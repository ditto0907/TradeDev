```python
import pandas as pd


csv_file_path = 'ES-12月14.csv' 

df = pd.read_csv(
    csv_file_path, 
    encoding='utf-8',       # 常用编码，可根据您的文件情况修改为 'gbk' 或 'latin-1'
    sep=',',                # 分隔符，默认是逗号，如果您的文件是 Tab 分隔，可改为 '\t'
    skiprows=0              # 跳过文件开头的行数 (0 表示不跳过)
)
df.info()
print("========================================================================================================")
print(df.head())
print("========================================================================================================")
print(df.tail())
print("========================================================================================================")
```
